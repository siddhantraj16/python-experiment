Question 10. check whether a number is armstrong or not

```python
num = int(input())
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** 3
    temp = temp // 10

if sum == num:
    print("Armstrong number")
else:
    print("Not Armstrong number")
```
