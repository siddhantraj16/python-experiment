Q1. Write a Python function to find the maximum and minimum numbers from a sequence of numbers. (Note: Do not use built-in functions.)

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum

nums = [10, 5, 25, 3, 18]
max_num, min_num = find_max_min(nums)

print("Maximum:", max_num)
print("Minimum:", min_num)
```

Q2.Write a Python function that takes a positive integer and returns the sum of the cube of
all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total = total + i**3
    return total
num = 5
result = sum_of_cubes(num)
print("Sum of cubes:", result)
```

Q3.Write a Python function to print 1 to n using recursion. (Note: Do not use loop)

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)
num = 5
print_numbers(num)
```

Q4.Write a recursive function to print Fibonacci series upto n terms.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)


def print_fibonacci(n, i=0):
    if i == n:
        return
    print(fibonacci(i), end=" ")
    print_fibonacci(n, i+1)
num = 6
print_fibonacci(num)
```

Q5.Write a lambda function to find volume of cone.

```python
import math

volume_cone = lambda r, h: (1/3) * math.pi * r * r * h
print("Volume:", volume_cone(3, 5))
```

Q6.Write a lambda function which gives tuple of max and min from a list.
Sample input: [10, 6, 8, 90, 12, 56]
Sample output: (90,6)

```python
find_max_min = lambda lst: (
    sorted(lst)[-1],
    sorted(lst)[0]
)
numbers = [10, 6, 8, 90, 12, 56]
result = find_max_min(numbers)
print(result)
```
