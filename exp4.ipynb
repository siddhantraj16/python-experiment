Q1. Write a program to count and display the number of capital letters in a given string

```python
s= input("enter a string :")
count = 0
for i in s:
  if i.isupper():
    count += 1
    print("No. of uppercase character :", count)
```

Q4. wap to enter a string and a substring . you have to print the number of times that the substring oocurs in the given string. string traversal will take place from left to right not from right to left

```python
string_main = input("enter a string : ")
sub_string = input("enter a substring : ")

count = 0
start = 0

while True:
    pos = string_main.find(sub_string, start)
    if pos == -1:
        break
    count = count + 1
    start = pos + 1

print("no. of occurrences :", count)
```

Q5.given a string containing both upper and lower case alphabets. write a python program to count the number of occurencies of each alphabet and display the same.

```python
s = input("Enter a string: ")
s = s.upper()

count = {}

for ch in s:
    if ch.isalpha():
        if ch in count:
            count[ch] += 1
        else:
            count[ch] = 1

for key in count:
    print(key, ":", count[key])
```

Q6. program to count number of unique words in a given sentence using sets.

Q7. create 2 sets s1 and s2 of n fruits each by taking input from user and find:
a) fruits which are in both sets s1 and s2
b) fruits only in s1 but not in s2
c) count all the fruits from s1 and s2

```python

n = int(input("Enter number of fruits in each set: "))


s1 = set()
print("Enter fruits for set 1:")
for _ in range(n):
    fruit = input().strip()
    s1.add(fruit)

s2 = set()
print("Enter fruits for set 2:")
for _ in range(n):
    fruit = input().strip()
    s2.add(fruit)

both = s1 & s2
print("Fruits in both sets:", both)

only_s1 = s1 - s2
print("Fruits only in set 1:", only_s1)

all_fruits = s1 | s2
print("Total number of distinct fruits:", len(all_fruits))
print("All distinct fruits:", all_fruits)
```

```python

n = int(input("Enter number of fruits: "))
s1 = set(input("Enter fruit: ") for _ in range(n))

s2 = set(input("Enter fruit: ") for _ in range(n))
print("Fruits in both sets:", s1 & s2)

print("Fruits only in set 1:", s1 - s2)

print("Total distinct fruits:", len(s1 | s2))
```
