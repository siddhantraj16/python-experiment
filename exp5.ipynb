Q1. scan n values in range 0-3 and print the number of times each value has occured

```python
n = int(input("Enter number of values: "))
count = [0, 0, 0, 0]
print("Enter values (0 to 3):")
for _ in range(n):
    value = int(input())
    if 0 <= value <= 3:
        count[value] += 1

for i in range(4):
    print(i, "occurred", count[i], "times")
```

Q2. create a tuple to store n numeric values and find average of all values

```python
num = int(input("Enter the number of values: "))
s = []

for i in range(num):
    n = int(input("Enter a number: "))
    s.append(n)
t = tuple(s)
avg = sum(t) / len(t)
print("The average of tuple items is:", avg)
```

Q4.Create a dictionary of n persons where key is name and value is city.
a) Display all names
b) Display all city names
c) Display student name and city of all students.
d) Count number of students in each city.

```python
# Create dictionary
n = int(input("Enter number of students: "))
students = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name] = city


print("Names are:")
for name in students:
    print(name)

print("Cities are:")
for city in students.values():
    print(city)


print("Student name and city:")
for name in students:
    print(name, "->", students[name])


print("Number of students in each city:")
count = {}

for city in students.values():
    count[city] = students.values().count(city)

for city in count:
    print(city, ":", count[city])
```

Q5.Store details of n movies in a dictionary by taking input from the user. Each movie must
store details like name, year, director name, production cost, collection made (earning)
& perform the following :-
a) print all movie details
b) display name of movies released before 2015
c) print movies that made a profit.
d) print movies directed by a particular director.

```python
# Input number of movies
n = int(input("Enter number of movies: "))
movies = {}

# Taking input
for i in range(n):
    name = input("\nEnter movie name: ")
    year = int(input("Enter release year: "))
    director = input("Enter director name: ")
    cost = float(input("Enter production cost: "))
    collection = float(input("Enter collection made: "))

    movies[name] = [year, director, cost, collection]

# a) Print all movie details
print("\nAll Movie Details:")
for name in movies:
    print(name, ":", movies[name])

# b) Movies released before 2015
print("\nMovies released before 2015:")
for name in movies:
    if movies[name][0] < 2015:
        print(name)

# c) Movies that made profit
print("\nMovies that made profit:")
for name in movies:
    if movies[name][3] > movies[name][2]:
        print(name)

# d) Movies by particular director
d = input("\nEnter director name to search: ")
print("Movies directed by", d, ":")

for name in movies:
    if movies[name][1] == d:
        print(name)
```

Q6.Create a Todo list Manager where users can add, view, and remove tasks. Use List for
storing tasks.

```python
tasks = []

while True:
    print("\n1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        task = input("Enter task: ")
        tasks.append(task)
        print("Task added!")

    elif choice == "2":
        print("Tasks:")
        for t in tasks:
            print(t)

    elif choice == "3":
        task = input("Enter task to remove: ")
        if task in tasks:
            tasks.remove(task)
            print("Task removed!")
        else:
            print("Task not found!")

    elif choice == "4":
        break

    else:
        print("Wrong choice!")
```

Q7.Create a contact book where users can store, search, update, and delete contacts. Use
dictionary for storing contacts.

```python
contacts = {}

while True:
    print("\n1.Add  2.Search  3.Update  4.Delete  5.View  6.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Name: ")
        phone = input("Phone: ")
        contacts[name] = phone
        print("Contact added!")

    elif choice == "2":
        name = input("Enter name to search: ")
        if name in contacts:
            print("Phone:", contacts[name])
        else:
            print("Not found!")

    elif choice == "3":
        name = input("Enter name to update: ")
        if name in contacts:
            phone = input("New phone: ")
            contacts[name] = phone
            print("Updated!")
        else:
            print("Not found!")

    elif choice == "4":
        name = input("Enter name to delete: ")
        if name in contacts:
            del contacts[name]
            print("Deleted!")
        else:
            print("Not found!")

    elif choice == "5":
        print("Contacts:", contacts)

    elif choice == "6":
        break

    else:
        print("Invalid choice!")
```
